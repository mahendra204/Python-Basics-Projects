```python
query = """
select * from (
SELECT 
    *, 
    RANK() OVER (partition by region ORDER BY total_revenue) AS revenue_rank
FROM (
    SELECT 
        delivery_partner, region,
        SUM(delivery_cost) AS total_revenue
    FROM delivery_logistics
    GROUP BY delivery_partner, region
) AS revenue_rankings )a where a.revenue_rank <= 3
"""

pd.read_sql(query, engine)
```
