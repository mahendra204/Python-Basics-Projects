```python
#select department, sum(salary) from table group by department -----department wise total salary

total_salary = employees3.groupby('department')['salary', 'id'].sum().reset_index(name=['total_salary_department_wise','total_id'])
```

```python
total_salary = employees3.groupby('department').agg(
    total_salary_department_wise=('salary', 'sum'),
    number_of_employees=('id', 'count'), 
    min_salary=('salary', 'min'), 
    max_salary=('salary', 'max')
).reset_index()
```
