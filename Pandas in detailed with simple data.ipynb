```python
def get_department(x):
    if x['age'] > 40:
        return 'Management'
    elif x['age'] > 30:
        return 'Sales'
    else:
        return 'Support'

df['department'] = df.apply(get_department, axis=1)
```

```python
df
```

```python
gdf=df.groupby('department')['salary'].sum().reset_index(name='total_salary_department_wise')
```

```python
gdf
```

```python
print(df.groupby('department')['salary'].count().reset_index(name='number_of_employees_dept_wise'))
```

```python
df.groupby('department')['salary'].mean().reset_index(name='avg_salary_dept_wise')
```
